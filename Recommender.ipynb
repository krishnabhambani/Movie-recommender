```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
import os
```

```python

movies = pd.read_csv('Data/tmdb_5000_movies.csv')
credits = pd.read_csv('Data/tmdb_5000_credits.csv')
```

```python
movies=movies.merge(credits, on= 'title')
```

```python
movies.shape
```

```python
movies.columns
```

```python
movies=movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.shape
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.shape
```

```python
movies.duplicated().sum()
```

```python
# Remove the brackets and all in dictionaries

import ast
def convert(text):
    l=[]
    for i in ast.literal_eval(text): #converting it into list
        l.append(i['name'])
    
    return l
```

```python
movies['genres'].apply(convert)
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
def convert_cast(text):
    l=[]
    counter=0
    for i in ast.literal_eval(text): #converting it into list
        if counter<3:
            l.append(i['name'])
        counter+=1
    
    return l
```

```python
movies['cast']=movies['cast'].apply(convert_cast)
```

```python
movies.head(2)
```

```python
#fetching the director of the movie

def fetch_director(text):
    l=[]
    for i in ast.literal_eval(text): #converting it into list
        if i['job']=='Director':
            l.append(i['name'])
            break
    return l
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies.head(2)
```

```python
movies['overview']=movies['overview'].apply(lambda x: x.split())
movies.head(2)
```

```python
# Sam Worthington - SamWorthington
def remove_space(word):
    l=[]
    for i in word:
        l.append(i.replace(" ",""))

    return l
```

```python
movies['cast']=movies['cast'].apply(remove_space)
movies['crew']=movies['crew'].apply(remove_space)
movies['genres']=movies['genres'].apply(remove_space)
movies['keywords']=movies['keywords'].apply(remove_space)
```

```python
movies['tag']=movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new_df=movies[['movie_id','title','tag']]
```

```python
new_df['tag']=new_df['tag'].apply(lambda x: " ".join(x))
```

```python
new_df.iloc[0]['tag']
```

```python
new_df['tag']=new_df['tag'].apply(lambda x: x.lower())
```

```python
import nltk
from nltk.stem import PorterStemmer
```

```python
ps=PorterStemmer()
```

```python
def stems(text):
    l=[]
    for i in text.split():
        l.append(ps.stem(i))
    
    return " ".join(l)
```

```python
new_df['tag']=new_df['tag'].apply(stems)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000, stop_words='english')
```

```python
vector= cv.fit_transform(new_df['tag']).toarray()
```

```python
vector.shape
vector
```

```python
# Cosine similarity

from sklearn.metrics.pairwise import cosine_similarity
similar=cosine_similarity(vector)
```

```python
similar
```

```python
new_df[new_df['title'] == 'Spider-Man'].index[0]
```

```python
def recommend(movie):
    index= new_df[new_df['title'] == movie].index[0]
    distance = sorted(list(enumerate(similar[index])), reverse=True , key=lambda x: x[1])

    for i in distance[1:6]:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Spider-Man')
```
